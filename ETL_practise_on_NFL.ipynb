```python
num_columns=orig_data.select_dtypes(include=["int64","float64"])
num_columns.shape
```

```python
Q1=num_columns.quantile(0.25)
Q3=num_columns.quantile(0.75)
IQR=Q3-Q1
upper=Q3+(1.5*IQR)
lower=Q1-(1.5*IQR)
```

```python
upper
```

```python
out_lier=num_columns[(num_columns>upper) | (num_columns<lower) ]
```

```python
outlier_mask = (num_columns > upper) | (num_columns < lower)
```

```python
outlier_mask.sum()
```

```python
med_data=num_columns.median()
d_cleaned=num_columns.mask(outlier_mask,med_data,axis=1)
```

```python
d_cleaned.shape
```

```python
d_cleaned.isna().sum().tolist()
```

```python
data=pd.concat([d_cleaned,orig_data.select_dtypes(object)],axis=1)
```
